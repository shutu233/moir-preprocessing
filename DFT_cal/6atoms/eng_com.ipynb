```python
import numpy as np
# force-field relaxation
dft_ini = -356.41827731
forc_local = [-356.42779100, -356.43268978, -356.43418881]
# dft-local start
dft_local = [-356.4283231395, -356.4332275859, -356.4347661121]

# dft_fire
dft_fire  = [-356.4283200880, -356.4332275859, -356.4347661121]

# turn
for i in range(0, 3):
    forc_local[i] = forc_local[i] - dft_ini
    forc_local[i] = forc_local[i] * 13.605698066819
    dft_local[i] = dft_local[i] - dft_ini
    dft_local[i] = dft_local[i] * 13.605698066819
    dft_fire[i] = dft_fire[i] - dft_ini
    dft_fire[i] = dft_fire[i] * 13.605698066819
```

```python
# plot 柱状图
import matplotlib.pyplot as plt
categories = ['AA', 'half', 'AB/BA']

bar_width = 0.25
index = np.arange(len(categories))

r1 = np.arange(len(categories))
r2 = [x for x in r1]
r3 = [x + bar_width for x in r2]

plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.size'] = 13    
plt.rcParams['font.weight'] = 'medium'

plt.bar(r2, dft_local, width=bar_width, label='DFT (w/ SW+KC start)', color='#90ee90')
plt.bar(r1, forc_local, width=bar_width, label='classical force field (SW+KC)', color='#ffb6c1')
plt.bar(r3, dft_fire, width=bar_width, label='DFT', color='#add8e6')

plt.ylabel('Potential Energy (eV)', weight='medium')
plt.xticks(index + bar_width/2, categories)
plt.legend(fontsize='13')
plt.savefig('plot.svg', format='svg')
plt.show()
```
